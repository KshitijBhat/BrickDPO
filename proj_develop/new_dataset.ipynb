```python
# Extract subset of data for 4 specific category_ids
import pandas as pd

# Define the category_ids to filter
category_ids = ["04379243", "03001627", "02958343", "04256520"]

# Load train dataset
print("Loading train dataset...")
df_train = pd.read_parquet('./datasets/stabletext2brick_train.parquet')
print(f"Original train dataset shape: {df_train.shape}")
print(f"Original train dataset columns: {df_train.columns.tolist()}")

# Filter train dataset
df_train_filtered = df_train[df_train['category_id'].isin(category_ids)]
print(f"\nFiltered train dataset shape: {df_train_filtered.shape}")
print(f"Category distribution in filtered train:")
print(df_train_filtered['category_id'].value_counts())

# Save filtered train dataset
output_train_path = './datasets/4class_stablebrick2text_train.parquet'
df_train_filtered.to_parquet(output_train_path, index=False)
print(f"\nSaved filtered train dataset to: {output_train_path}")

# Load test dataset
print("\n" + "="*50)
print("Loading test dataset...")
df_test = pd.read_parquet('./datasets/stabletext2brick_test.parquet')
print(f"Original test dataset shape: {df_test.shape}")
print(f"Original test dataset columns: {df_test.columns.tolist()}")

# Filter test dataset
df_test_filtered = df_test[df_test['category_id'].isin(category_ids)]
print(f"\nFiltered test dataset shape: {df_test_filtered.shape}")
print(f"Category distribution in filtered test:")
print(df_test_filtered['category_id'].value_counts())

# Save filtered test dataset
output_test_path = './datasets/4class_stablebrick2text_test.parquet'
df_test_filtered.to_parquet(output_test_path, index=False)
print(f"\nSaved filtered test dataset to: {output_test_path}")

print("\n" + "="*50)
print("Done! Both filtered datasets have been saved.")
```
